## Technical intermission!

### What went wrong?

**Recap of linear models**

To understand, let's recap the linear model.
Suppose $y$ refers to whether an O-ring failed or not
and let $x$ be temperature at launch.

We have observations $(x_i,y_i)_{i=1}^{23}$ and are modeling,
for $i=1,\dots,23$,

$$
    Y_i = \beta_0 + \beta_1 x_i + \epsilon_i, \qquad \epsilon_i \sim \mathcal{N}(0,\sigma^2),
$$

where the errors are independent from one another and 
$\beta_0,\beta_1, \sigma^2$ are parameters to be estimated.


**A different perspective.**

The previous formulation makes it look like the randomness comes from the errors,
when in fact it comes from the response variable $Y_i$.

We can instead think of it as modeling the mean of the response, which itself is Gaussian:

$$
    Y_i \sim \mathcal{N} (\mu_i, \sigma^2), \\
    \mu_i = \beta_0+\beta_1 x_i.
$$

Although equivalent, this new formulation makes a clear separation between: 
- the *random component*, $Y_i \sim \mathcal{N} (\mu_i, \sigma^2)$, and
- the *systematic component*, $\mu_i = \beta_0+\beta_1 x_i$.



**What went wrong?**

By thinking of linear models in terms of their components, 
we can identify precisely where our model failed:
- Normality is not satisfied
- the output Y might fall outside the range of 0 and 1. 

### Logistic regression

Let's fix the first issue.
Our response is binary, thus cannot be distributed Gaussian,
but instead Bernoulli: $Y_i \sim \mathrm{Bernoulli}(p_i)$.


Following the random/systematic component framework, we can write

$$
    Y_i \sim \mathrm{Bernoulli}(p_i),\\
    p_i = \beta_0 + \beta_1 x_i.
$$

**Problem:** We can't equate $p_i \in (0,1)$ with $\beta_0 + \beta_1 x_i \in \mathbb{R}$.
This is the second limitation of the linear model:
it assumes that the response mean and the predictor live in the same space.
To address this, we transform the response mean to an appropriate space.
One possibility&mdash;which results in the model called *logistic regression*&mdash;is
using the logit function:

$$
    Y_i \sim \mathrm{Bernoulli}(p_i),\\
    \log \frac{p_i}{1-pi} = \beta_0 + \beta_1 x_i.
$$


In this case, the logit function is called a *link function*. 
The logit link is invertible. We can also write

$$
    Y_i \sim \mathrm{Bernoulli}(p_i),\\
    p_i = \frac{1}{1+\exp(-(\beta_0 + \beta_1 x_i))} 
    = \frac{\exp(\beta_0 + \beta_1 x_i)}{1+\exp(\beta_0 + \beta_1 x_i)}
$$

All of these are equivalent and called *logistic regression*.


**The likelihood.**

Recall that the Bernoulli likelihood is given by

$$
  p^y (1-p)^{1-y},
$$

and so the likelihood of the model parameters given our data is

$$
    L(\beta_0,\beta_1; x,y) = \prod_{i=1}^{23} p_i^{y_i} (1-p_i)^{1-y_i}.
$$

Plugging in $p_i$ as a function of the model parameters results in

$$
    L(\beta_0,\beta_1; x,y) = \prod_{i=1}^{23} 
    \left( \frac{1}{1+\exp(-(\beta_0 + \beta_1 x_i))} \right)^{y_i}
    \left( \frac{1}{1+\exp(\beta_0 + \beta_1 x_i)} \right)^{1-y_i}.
$$


**Fitting the model.**

To find the maximum likelihood estimator,
we have to maximize the likelihood.
It turns out that, in this case, there is no closed-form solution, due to the nonlinearity of the logistic sigmoid function.

But, not all is lost.
Even though we can't write down a formula for the MLE,
we can still find it in practice using one of many optimization algorithms,
such as gradient descent or Newton's method.

Further, it is possible to show that the likelihood is concave.
This means that the MLE exists and is unique,
and that most optimization algorithms should be able to find it.

## Fitting a logistic regression to the Challenger data

We use the `glm` function in R.

```r
logit_mod <- glm(distress ~ launch_temp, family = "binomial", data = challenger)
summary(logit_mod)
```

The summary gives us the estimates,
and it also prints out information about the optimization algorithm used to find the MLE.

Now let's plot the estimated probability of failure as a function of temperature, i.e.,

$$
    \hat{p_i} = \frac{1}{1+\exp(-(\hat{\beta}_0 + \hat{\beta}_1 x_i))}.
$$

```r
challenger %>% 
    ggplot(aes(x = launch_temp, y = distress)) +
    geom_point(size=4) +
    theme(text = element_text(size=25)) +
    scale_y_continuous(breaks=0:1) +
    labs(x="Launch temperature (°C)",
        y = "O-ring failure status") +
    geom_smooth(method = "glm", formula = y~x, method.args = list(family = "binomial"), se = FALSE)
```

We can also predict the probability of a failure ocurring at -0.6°C
just like we would in a linear regression.

```r
eta = function(x) logit_mod$coefficients[1] + logit_mod$coefficients[2]*x
print(paste0("Probability of failure at -0.6°C: ", 1/(1+exp(-eta(-0.6)))))
```

Or use `predict.glm` in R (preferred especially when predicting multiple values).

```r
predict.glm(logit_mod, newdata = data.frame(launch_temp = -0.6), type="response")
```

### Measuring goodness-of-fit

The p-values are a good starting place,
but as we saw earlier that might not paint the full picture.

**Deviance.**

Measures deviation from the "perfect" or saturated model, in which

$$
    p_i = y_i, \qquad i=1,\dots,23.
$$

What is the likelihood of this model? 
Assuming for the sake of simplicity that $0^0=1$,

$$
    L_s = \prod_{i=1}^{23} y_i^{y_i} (1-y_i)^{1-y_i} = 1.
$$

The deviance of a given model with parameters $\beta_0,\beta_1$ is defined as

$$
    D(\beta_0, \beta_1) = -2 (\log L(\beta_0, \beta_1; x,y) - \log L_s) = -2 \log L(\beta_0, \beta_1; x,y).
$$

From R's output,
- null deviance refers to $D(\hat{\beta}_0, 0)$ (null model, only intercept) and
- residual deviance refers to $D(\hat{\beta}_0, \hat{\beta}_1)$.

Let's print the summary of the logistic model again:

```r
summary(logit_mod)
```

**Confusion matrix.**

For classification, this is a good way to evaluate the model's performance.

Idea: predict the probability of failure for your data, then classify data as predicted failure or success.

We first need to set a threshold for predicting a failure, say 0.5,
but this value can be modified depending on the setting.

```r
threshold <- 0.5

challenger <- 
    challenger %>% 
    dplyr::mutate(phat = predict.glm(logit_mod, type="response"),
                 prediction = as.integer(phat>threshold)) %>% 
    dplyr::select(launch_temp, distress, prediction, phat)
challenger
```

A confusion matrix summarizes the distress and prediction columns above.
There are many metrics used in practice.
[Wikipedia](https://en.wikipedia.org/wiki/Confusion_matrix) has a good article on it.

```r
library(caret)
caret::confusionMatrix(factor(challenger$prediction), factor(challenger$distress))
```

**Value-prediction plot.**

Useful to make sure predictions are not way off.

On the x-axis you plot your predicted probability and 
on the y-axis you plot the true value of the response variable.

```r
challenger %>% 
    ggplot(aes(x = phat, y = distress)) +
    geom_point(size=4) +
    theme(text = element_text(size=25)) +
    scale_y_continuous(breaks=0:1) +
    labs(x = expression(hat(p)),
        y = "O-ring failure status")
```

In general,
you want to have many points in the lower-left and upper-right corners,
and few points in the other two corners.

## Generalized linear models

These are a generalization of linear models (duh) due to [Nelder and Wedderburn (1972)](https://www.jstor.org/stable/2344614?origin=crossref&seq=1#metadata_info_tab_contents)
(but also see
[McCullagh and Nelder (1989](http://www.utstat.toronto.edu/~brunner/oldclass/2201s11/readings/glmbook.pdf)).
Logistic regression is a GLM, as are many other models.
The definition builds on the random and systemic components we discussed before.


**Random component.**

As before, we assume that the observations $y_1,\dots,y_N$ are realizations of a random variable.
The only condition here is that the distribution of the response variable $Y_i$
has to be in the [exponential family of distributions](https://en.wikipedia.org/wiki/Exponential_family).

Without getting into details,
it turns out that many popular distributions are in the exponential family:
- Bernoulli,
- Binomial,
- Gaussian,
- Poisson,
- Exponential,
- Gamma,
- Beta,
- many others!


**Systematic component.**

Here we also have the same basic principle.
If $\mu_i$ is the expectation of $Y_i$ (given the data!) then we try to estimate it
using a linear combination of covariates, $\beta_0+\beta_1 x_i$.

To ensure that we can equate $\mu_i$ with $\beta_0+\beta_1 x_i$,
we need a *link function* $g$ such that

$$
    g(\mu_i) = \beta_0+\beta_1 x_i, \qquad i=1,\dots,N.
$$

The link function is not unique,
and indeed different choices lead to different models.


**GLMs.**

A GLM can thus be specified as:

$$
    Y_i \sim f, \qquad f \text{ in exponential family with mean } \mu_i, \\
    g(\mu_i) = \beta_0 + \beta_1 x_i.
$$

Here are some popular combinations of random components and link functions:


| distribution of $Y_i$       | mean $\mu_i$ | link $g(\mu)$            | name                  |commonly used for|
|-----------------------------|--------------|--------------------------|-----------------------|-----------------|
| $\mathcal{N}(\mu,\sigma^2)$ | $mu$         | $\mu$                    | linear regression     |continuous outcomes|
| Bernoulli$(p)$              | $p$          | $\log \frac{\mu}{1-\mu}$ | logistic regression   | binary outcome|
| Poisson$(\lambda)$          | $\lambda$    | $\log(\mu)$              | Poisson regression    | count data; with careful construction it can also model rate; overdispersion might occur|
| Gamma$(a,b)$                | $a/b$        | $-1/\mu$                 | Gamma regression      | time-to-event data|
